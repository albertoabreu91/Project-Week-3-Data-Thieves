```python
refugees_raw = pd.read_csv(file_path, skiprows=4 )
```

```python
refugees_raw
```

There are some years that have no data. This is because the World Bank Data was founded in 1960, but they started collecting data about the refugees on 1990. Hence, we can drop those columns. 

```python
drop_years = [str(i) for i in range(1960,1991)]
```

```python
refugees_raw = refugees_raw.drop(drop_years, axis=1)
```

```python
refugees_raw
```

We will drop more columns that are not of our interest, such as: 
    'Indicator name' and 'Indicator Code'

```python
refugees_raw = refugees_raw.drop(['Indicator Name', 'Indicator Code'], axis=1)
```

```python
refugees_raw.head()
```

### Finding missing values in our dataset

We are looking for the NaN values in our dataset, to see what we can do with them in order to perform the analysis. 

```python
refugees_raw.info()
```

From this table we can see that there are some values that are none, but taking a look at the data we can see that there are not any 0. In this case, we will replace the NaN with zeros to be able to plot the evolution of the refugees year by year. 

However, it should be noticed that there can be a bias since there are some countries where all the elements are 0. 

Furthermore, the data of the first years of the studies might not be accurated at all, and since the world Bank data is retrieving more data, there are less NaN values year by year.

```python
refugees = refugees_raw.fillna(0) 
```
