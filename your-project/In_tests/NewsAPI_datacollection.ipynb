## 3. News collection

Let's create a function that allows us to get data from the API, considering that is has some pagination and each page retrieves 15 news with the topic we have included. 

*To remember: There's a limitation on the number of requests we can do. NewsAPI limits the requests to 5 pages per topic.*

```python
import datetime
```

```python
# Function to access the api
def newsAPI_getnews(topic, from_page, to_page):
    all_news_lst = []
    for page in range(from_page,to_page+1):
        news_request = newsapi.get_everything(q=topic, sort_by='relevancy', page=page )
        news_df = pd.DataFrame(news_request["articles"])
        all_news_lst.append(news_df)
        print(datetime.datetime.now())
    return pd.concat(all_news_lst).reset_index(drop=True) 
```

```python
refugees = newsAPI_getnews(topic='refugees', from_page= 1, to_page=5)
```

```python
refugees
```

### 3.1. NewsAPI Data Wrangling and cleaning

If we take a look at the dataframe, we can see some insights or things that we don't need, or that we might need to change.

1. Change "PublishedAt" to datetime. 
2. See if content and description have the same content inside and one can be removed
3. Split 'source' column, that contains one dictionary, to two new columns: source_id and source_name
4. URLtoImage column can be drop out
5. Renaming of the columns to follow the standards on databases naming "snake_case"

```python
def NewsAPI_format_df(news_df):
    news_df['publishedAt'] = pd.to_datetime(news_df['publishedAt'])
    news_df_format = news_df.copy()
    news_df_format['source_id'] = news_df['source'].apply(lambda x: x['id'])
    news_df_format['source_name'] = news_df['source'].apply(lambda x: x['name'])
    news_df_format = news_df_format.drop(['content','source','urlToImage'], axis=1)
    news_df_format.columns = ['author', 'description', 'date_published', 'title', 'url', 'source_name_id', 'source_name']
    
    return news_df_format
```
